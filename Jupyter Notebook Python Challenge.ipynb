```python
# Working out m=c^2 vequation to start mapping the letters:

# To work out c^2, take the numeral position of c in the alphabet and square it:
# c=3rd position in alphabet
# 3^2=9
# The 9th letter in the alphabet = i 

# Therefore m=i or (x-4)
# i.e (m,i) (n,j) (o,k) (p,l) etc... 


  
```

```python
# Enter Raw data
raw = "lxxtw://m.mqkjpmt.gsq/3conkm.ntk"

# (x-4) m=i, n=j, 
a = "mnopqrstuvwxyzabcdefghijkl,. '()"
b = "ijklmnopqrstuvwxyzabcdefgh,. '()"
```

```python
# Add a key-value pair for the ':, /, 3' characters
b += ':'
a += ':'
b += '/'
a += '/'
b += '3'
a += '3'
```

```python
# Create the dictionary
char_map = dict(zip(a, b))
```

```python
print("Drum roll! And your prize is! .... a link.... of a universal feeling...")

# Use the dictionary to map each character in the raw variable
print("".join([char_map[x] for x in raw]))

# The URL answer below 
print("... enjoy.")
```
